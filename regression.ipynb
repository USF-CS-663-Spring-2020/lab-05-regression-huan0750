# Toluca

```python
import pandas as pd
```

```python
df=pd.read_csv('toluca.csv')
df
```

```python
df.plot(kind='scatter', x='lotSize', y='workHours')
```

```python
import seaborn as sns
plot=sns.lmplot(y='workHours', x='lotSize', data=df)
```

```python
rl=sns.residplot('lotSize','workHours',data=df)
rl.set(xlabel='Lot Size', ylabel='residuals')
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_t = LinearRegression()
lin_reg_t.fit(df.drop('workHours',axis=1),df.iloc[:,1])
lin_reg_t
print(lin_reg_t.coef_)
print(lin_reg_t.intercept_)
```
